```python
polarizacion = pd.read_csv('/content/resumen_polarizacion_chile.csv')

polarizacion.columns = polarizacion.columns.str.strip()

polarizacion
```

```python
polarizacion['anio'] = polarizacion['año_eleccion'].astype(str).str[:4].astype(int)

polarizacion[['año_eleccion', 'anio', 'score_final']]
```

```python
plt.figure(figsize=(10,6))

plt.plot(
    polarizacion['anio'],
    polarizacion['score_final'],
    marker='o'
)

plt.title('Evolución de la polarización en elecciones presidenciales')
plt.xlabel('Año')
plt.ylabel('Score de polarización')
plt.xticks(polarizacion['anio'], rotation=45)
plt.grid(True)

plt.show()
```

```python
polarizacion_vs_nulos = pd.merge(
    polarizacion[['anio', 'score_final']],
    votos_por_eleccion_limpio[votos_por_eleccion_limpio['vuelta'] == 1][['anio', 'porcentaje_nulos']],
    on='anio',
    how='inner'
)

polarizacion_vs_nulos = polarizacion_vs_nulos.sort_values('anio')

polarizacion_vs_nulos
```

```python
plt.figure(figsize=(12,6))

plt.plot(
    polarizacion_vs_nulos['anio'],
    polarizacion_vs_nulos['score_final'],
    marker='o',
    label='Polarización'
)

plt.plot(
    polarizacion_vs_nulos['anio'],
    polarizacion_vs_nulos['porcentaje_nulos'],
    marker='o',
    label='Voto nulo (%)'
)

plt.title('Evolución de polarización y voto nulo en primera vuelta')
plt.xlabel('Año')
plt.ylabel('Valor')
plt.xticks(polarizacion_vs_nulos['anio'], rotation=45)
plt.legend()
plt.grid(True)

plt.show()
```

```python
polarizacion_vs_nulos_normalizado = polarizacion_vs_nulos.copy()

polarizacion_vs_nulos_normalizado['polarizacion_normalizada'] = (
    polarizacion_vs_nulos_normalizado['score_final'] /
    polarizacion_vs_nulos_normalizado['score_final'].max()
) * 100

polarizacion_vs_nulos_normalizado['nulos_normalizados'] = (
    polarizacion_vs_nulos_normalizado['porcentaje_nulos'] /
    polarizacion_vs_nulos_normalizado['porcentaje_nulos'].max()
) * 100

polarizacion_vs_nulos_normalizado
```

```python
plt.figure(figsize=(12,6))

plt.plot(
    polarizacion_vs_nulos_normalizado['anio'],
    polarizacion_vs_nulos_normalizado['polarizacion_normalizada'],
    marker='o',
    label='Polarización normalizada'
)

plt.plot(
    polarizacion_vs_nulos_normalizado['anio'],
    polarizacion_vs_nulos_normalizado['nulos_normalizados'],
    marker='o',
    label='Voto nulo normalizado'
)

plt.title('Evolución comparada de polarización y voto nulo')
plt.xlabel('Año')
plt.ylabel('Índice normalizado, máximo = 100')
plt.xticks(polarizacion_vs_nulos_normalizado['anio'], rotation=45)
plt.legend()
plt.grid(True)

plt.show()
```
