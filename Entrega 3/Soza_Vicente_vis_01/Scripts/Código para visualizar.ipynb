```python
import pandas as pd

df = pd.read_csv("Base de Datos Elecciones y Candidatos.csv", sep=';')
df.columns = df.columns.str.strip().str.lower()
```

```python
df['total_candidato'] = df.loc[:, 'tarapacá':'magallanes'].sum(axis=1)
```

```python
nulos_total = df['nulos de la votación'].sum()

df_nulos = pd.DataFrame({
    'candidatos': ['Nulos'],
    'total_candidato': [nulos_total]
})
```

```python
df_plot = pd.concat([
    df[['candidatos', 'total_candidato']],
    df_nulos
])
```

```python
import altair as alt

grafico = alt.Chart(df_plot).mark_bar().encode(
    x=alt.X('candidatos:N', sort='-y', title='Candidatos'),
    y=alt.Y('total_candidato:Q', title='Votos'),
    color='candidatos:N',
    tooltip=['candidatos', 'total_candidato']
).properties(
    title='Comparación de votos: candidatos vs nulos'
)

grafico
```

```python
text = grafico.mark_text(
    align='center',
    dy=-5
).encode(
    text='total_candidato:Q'
)

grafico + text
```

```python
grafico.save("grafico.html")
```
