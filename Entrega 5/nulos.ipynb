```python
comparacion_vueltas = nacional_grafico.copy()

tabla_salto = comparacion_vueltas.pivot_table(
    index="anio",
    columns="vuelta",
    values="pct_nulos_blancos",
    aggfunc="first"
).reset_index()

tabla_salto = tabla_salto.rename(columns={
    1: "primera_vuelta",
    2: "segunda_vuelta"
})

tabla_salto = tabla_salto.dropna(subset=["primera_vuelta", "segunda_vuelta"])

tabla_salto["diferencia_pp"] = tabla_salto["segunda_vuelta"] - tabla_salto["primera_vuelta"]

tabla_salto["direccion"] = np.where(
    tabla_salto["diferencia_pp"] >= 0,
    "Sube en segunda vuelta",
    "Baja en segunda vuelta"
)

display(tabla_salto)
```

```python
polarizacion = pd.read_csv("polarizacion_elecciones_chile_1989_2025.csv")

polarizacion.columns = polarizacion.columns.str.strip()

display(polarizacion.head())
print(polarizacion.columns.tolist())
```

```python
polarizacion = polarizacion.rename(columns={
    "Año": "anio",
    "Vuelta": "vuelta",
    "Polarización vuelta 1": "polarizacion_v1",
    "Polarización vuelta 2": "polarizacion_v2",
    "Polarización total": "polarizacion_total",
    "Polarizacion vuelta 1": "polarizacion_v1",
    "Polarizacion vuelta 2": "polarizacion_v2",
    "Polarizacion total": "polarizacion_total"
})

polarizacion.columns = polarizacion.columns.str.strip()

display(polarizacion.head())
print(polarizacion.columns.tolist())
```

```python
for col in ["anio", "vuelta", "polarizacion_v1", "polarizacion_v2", "polarizacion_total"]:
    if col in polarizacion.columns:
        polarizacion[col] = (
            polarizacion[col]
            .astype(str)
            .str.replace(",", ".", regex=False)
            .replace("nan", np.nan)
        )
        polarizacion[col] = pd.to_numeric(polarizacion[col], errors="coerce")

display(polarizacion.head())
```

```python
polarizacion_resumen = (
    polarizacion
    .groupby("anio", as_index=False)
    .agg(
        polarizacion_v1=("polarizacion_v1", "max"),
        polarizacion_v2=("polarizacion_v2", "max"),
        polarizacion_total=("polarizacion_total", "max")
    )
)

display(polarizacion_resumen)
```

```python
polarizacion_grafico = polarizacion_resumen.copy()

polarizacion_grafico = polarizacion_grafico.sort_values("anio")

polarizacion_larga = polarizacion_grafico.melt(
    id_vars="anio",
    value_vars=["polarizacion_v1", "polarizacion_v2"],
    var_name="vuelta_tipo",
    value_name="polarizacion"
)

polarizacion_larga["vuelta_label"] = polarizacion_larga["vuelta_tipo"].map({
    "polarizacion_v1": "Primera vuelta",
    "polarizacion_v2": "Segunda vuelta"
})

polarizacion_larga = polarizacion_larga.dropna(subset=["polarizacion"])

display(polarizacion_larga)
```

```python
grafico_5 = alt.Chart(polarizacion_larga).mark_bar(
    cornerRadiusEnd=4
).encode(
    x=alt.X(
        "anio:O",
        title="Año",
        axis=alt.Axis(labelAngle=0)
    ),
    y=alt.Y(
        "polarizacion:Q",
        title="Puntaje de polarización",
        scale=alt.Scale(domain=[0, 5])
    ),
    color=alt.Color(
        "vuelta_label:N",
        title=None,
        scale=alt.Scale(
            domain=["Primera vuelta", "Segunda vuelta"],
            range=["#64748b", "#d94841"]
        )
    ),
    xOffset=alt.XOffset("vuelta_label:N"),
    tooltip=[
        alt.Tooltip("anio:O", title="Año"),
        alt.Tooltip("vuelta_label:N", title="Vuelta"),
        alt.Tooltip("polarizacion:Q", title="Polarización", format=".1f")
    ]
).properties(
    title={
        "text": "La temperatura política sube en las últimas presidenciales",
        "subtitle": "Puntaje de polarización de la oferta presidencial por primera y segunda vuelta",
        "anchor": "start",
        "fontSize": 18,
        "subtitleFontSize": 12,
        "subtitleColor": "#374151"
    },
    width=760,
    height=400
).configure_view(
    strokeWidth=0
).configure_axis(
    gridColor="#e5e7eb",
    labelFontSize=11,
    titleFontSize=12
).configure_legend(
    orient="top",
    title=None
)

grafico_5
```

```python
polarizacion_cruce = pd.merge(
    nacional_grafico[['anio', 'vuelta_label', 'pct_nulos_blancos']],
    polarizacion_larga[['anio', 'vuelta_label', 'polarizacion']],
    on=['anio', 'vuelta_label'],
    how='outer'
)
cruce_linea = polarizacion_cruce.copy()

cruce_linea["pct_nulos_blancos_norm"] = (
    cruce_linea["pct_nulos_blancos"] / cruce_linea["pct_nulos_blancos"].max() * 5
)

cruce_linea_larga = cruce_linea.melt(
    id_vars=["anio", "vuelta_label"],
    value_vars=["polarizacion", "pct_nulos_blancos_norm"],
    var_name="variable",
    value_name="valor"
)

cruce_linea_larga["variable_label"] = cruce_linea_larga["variable"].map({
    "polarizacion": "Polarización",
    "pct_nulos_blancos_norm": "Nulos + blancos normalizados"
})

display(cruce_linea_larga)
```

```python
avance_anual = (
    nacional_grafico
    .groupby("anio", as_index=False)
    .agg(
        nulos_total=("nulos_total", "sum"),
        blancos_total=("blancos_total", "sum"),
        nulos_blancos_total=("nulos_blancos_total", "sum"),
        votos_emitidos=("votos_emitidos", "sum")
    )
)

avance_anual["pct_nulos"] = avance_anual["nulos_total"] / avance_anual["votos_emitidos"] * 100
avance_anual["pct_blancos"] = avance_anual["blancos_total"] / avance_anual["votos_emitidos"] * 100
avance_anual["pct_nulos_blancos"] = avance_anual["nulos_blancos_total"] / avance_anual["votos_emitidos"] * 100

avance_anual = avance_anual.sort_values("anio")

display(avance_anual)
```

```python
regional_grafico = regional.copy()

regional_grafico["anio"] = pd.to_numeric(regional_grafico["anio"], errors="coerce")
regional_grafico["vuelta"] = pd.to_numeric(regional_grafico["vuelta"], errors="coerce")
regional_grafico["pct_nulos_blancos"] = pd.to_numeric(
    regional_grafico["pct_nulos_blancos"],
    errors="coerce"
)

regional_grafico["vuelta_label"] = regional_grafico["vuelta"].map({
    1: "Primera vuelta",
    2: "Segunda vuelta"
})

regional_grafico = regional_grafico.dropna(
    subset=["anio", "vuelta", "region", "pct_nulos_blancos"]
)

regional_grafico = regional_grafico.sort_values(["region", "vuelta", "anio"])

display(regional_grafico.head())
print(regional_grafico.shape)
print(regional_grafico["region"].nunique())
print(regional_grafico["region"].unique())
```

```python
grafico_regional = alt.Chart(regional_grafico, width=180, height=130).mark_line(
    point=True,
    strokeWidth=2
).encode(
    x=alt.X(
        "anio:Q",
        title="Año",
        scale=alt.Scale(domain=[1989, 2025]),
        axis=alt.Axis(
            values=[1989, 1993, 1999, 2005, 2009, 2013, 2017, 2021, 2025],
            format="d",
            labelAngle=0
        )
    ),
    y=alt.Y(
        "pct_nulos_blancos:Q",
        title="% nulos + blancos",
        axis=alt.Axis(format=".1f")
    ),
    color=alt.Color(
        "vuelta_label:N",
        title=None,
        scale=alt.Scale(
            domain=["Primera vuelta", "Segunda vuelta"],
            range=["#64748b", "#d94841"]
        )
    ),
    detail="vuelta_label:N",
    tooltip=[
        alt.Tooltip("region:N", title="Región"),
        alt.Tooltip("anio:Q", title="Año", format="d"),
        alt.Tooltip("vuelta_label:N", title="Vuelta"),
        alt.Tooltip("pct_nulos_blancos:Q", title="% nulos + blancos", format=".2f"),
        alt.Tooltip("nulos_blancos:Q", title="Nulos + blancos", format=","),
        alt.Tooltip("votos_emitidos:Q", title="Votos emitidos", format=",")
    ]
).properties(
    title={
        "text": "El voto nulo y blanco no evoluciona igual en todas las regiones",
        "subtitle": "Porcentaje de votos nulos y blancos por región en elecciones presidenciales chilenas, 1989–2025",
        "anchor": "start",
        "fontSize": 18,
        "subtitleFontSize": 12,
        "subtitleColor": "#374151"
    }
).facet(
    facet="region:N",
    columns=4
).configure_view(
    strokeWidth=0
).configure_axis(
    gridColor="#e5e7eb",
    labelFontSize=9,
    titleFontSize=10
).configure_header(
    labelFontSize=11,
    labelFontWeight="bold"
).configure_legend(
    orient="top",
    title=None
)

grafico_regional
```

```python
import os
import shutil
import unicodedata

os.makedirs("graficos_html", exist_ok=True)

# -----------------------------
# 1. GUARDAR GRÁFICOS PRINCIPALES
# -----------------------------

graficos_principales = {
    "01_evolucion_nacional_nulos_blancos.html": grafico_1,
    "02_diferencia_primera_segunda_vuelta.html": grafico_2,
    "03_sistema_electoral_nulos_blancos.html": grafico_3,
    "05_polarizacion_oferta_presidencial.html": grafico_5,
    "07_heatmap_regional_nulos_blancos.html": grafico_heatmap_regional,
    "08_ranking_regional_2025_nulos_blancos.html": grafico_ranking_2025,
    "11_mapa_calor_chile_2025_nulos_blancos.html": grafico_mapa_chile_2025,
    "12_cartograma_chile_todos_los_anios_nulos_blancos.html": grafico_mapa_chile_todos
}

for nombre_archivo, grafico in graficos_principales.items():
    grafico.save(f"graficos_html/{nombre_archivo}")
    print("Guardado:", nombre_archivo)


# -----------------------------
# 2. FUNCIÓN PARA NOMBRAR REGIONES
# -----------------------------

def limpiar_nombre_region(nombre):
    nombre = str(nombre).lower()
    nombre = nombre.replace("ñ", "n")
    nombre = nombre.replace("’", "")
    nombre = nombre.replace("'", "")
    nombre = unicodedata.normalize("NFKD", nombre)
    nombre = "".join(c for c in nombre if not unicodedata.combining(c))
    nombre = nombre.replace(" ", "_")
    return nombre


# -----------------------------
# 3. GUARDAR GRÁFICOS REGIONALES
# -----------------------------

regiones = sorted(regional_grafico["region"].dropna().unique())

for region_elegida in regiones:
    datos_region = regional_grafico[
        regional_grafico["region"] == region_elegida
    ].copy()

    grafico_region = alt.Chart(datos_region).mark_line(
        point=True,
        strokeWidth=3
    ).encode(
        x=alt.X(
            "anio:Q",
            title="Año",
            scale=alt.Scale(domain=[1989, 2025]),
            axis=alt.Axis(
                values=[1989, 1993, 1999, 2005, 2009, 2013, 2017, 2021, 2025],
                format="d",
                labelAngle=0
            )
        ),
        y=alt.Y(
            "pct_nulos_blancos:Q",
            title="% nulos + blancos",
            axis=alt.Axis(format=".1f")
        ),
        color=alt.Color(
            "vuelta_label:N",
            title=None,
            scale=alt.Scale(
                domain=["Primera vuelta", "Segunda vuelta"],
                range=["#64748b", "#d94841"]
            )
        ),
        detail="vuelta_label:N",
        tooltip=[
            alt.Tooltip("region:N", title="Región"),
            alt.Tooltip("anio:Q", title="Año", format="d"),
            alt.Tooltip("vuelta_label:N", title="Vuelta"),
            alt.Tooltip("pct_nulos_blancos:Q", title="% nulos + blancos", format=".2f"),
            alt.Tooltip("nulos_blancos:Q", title="Nulos + blancos", format=","),
            alt.Tooltip("votos_emitidos:Q", title="Votos emitidos", format=",")
        ]
    ).properties(
        title={
            "text": f"Evolución del voto nulo y blanco en {region_elegida}",
            "subtitle": "Porcentaje de votos nulos y blancos en elecciones presidenciales, 1989–2025",
            "anchor": "start",
            "fontSize": 18,
            "subtitleFontSize": 12,
            "subtitleColor": "#374151"
        },
        width=820,
        height=420
    ).configure_view(
        strokeWidth=0
    ).configure_axis(
        gridColor="#e5e7eb",
        labelFontSize=11,
        titleFontSize=12
    ).configure_legend(
        orient="top",
        title=None
    )

    nombre_archivo = "region_" + limpiar_nombre_region(region_elegida) + ".html"
    grafico_region.save(f"graficos_html/{nombre_archivo}")
    print("Guardado:", nombre_archivo)


# -----------------------------
# 4. CREAR ZIP FINAL
# -----------------------------

shutil.make_archive("graficos_html", "zip", "graficos_html")

print("LISTO: se creó graficos_html.zip")
print("Descarga graficos_html.zip desde el panel izquierdo de Colab.")
```
